```python
def todict(data_class):
    for word in data_class:
        # selecteer dataset
        column = dataset[word]
        counter = 0 

        # loop door rijen van participant_name kolom
        for row in column:    

        #     als rij niet leeg is (soms is ie leeg)
            if not pd.isnull(row):
                if '||' in row:
                    cleanrow = row.split("||")
                elif '|' in row:
                    cleanrow = row.split('|')
                else:
                    cleanrow = row.split()

        #         maak er een dictonary van
#                 print(row)
                splitted_dict = {}
                for element in cleanrow:
                    if '::' in element:
                        index, status = element.split("::")
                    elif ':' in element:
                        index, status = element.split(":")    
                    splitted_dict[index] = status

                dataset.at[counter, word] = splitted_dict
            counter += 1
#             print(dataset[word][counter])
    
todict(['participant_name', 'gun_stolen', 'gun_type', 'participant_age', 'participant_age_group', 'participant_gender', 'participant_status', 'participant_type', 'participant_relationship' ])
# todict(['participant_name'])
# todict(['gun_stolen'])
# todict(['gun_type'])


dataset

    
```
