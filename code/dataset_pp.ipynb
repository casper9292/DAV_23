```python
# Load libraries
import numpy as np
import pandas as pd
import json
import csv
```

```python
##############################
# clean up empty lines in csv
##############################

with open("dataset.csv") as fIn, open("dataset_clean.csv", "w") as fOut:
    
    counter = 0   
    for line in fIn:
        nextline = line.strip('\n')
        
        if counter == 0:
            lastline = line
          
        if counter > 0:
            # if nextline starts with incident_id, this line has all columns
            if nextline[0:5].isdigit():
                fOut.write(lastline)
                fOut.write('\n')
                lastline = nextline

            # else add the lines together
            else:
                lastline = lastline + nextline
        
        counter = counter + 1
```

```python
# read csv
dataset = pd.read_csv("dataset_clean.csv", sep=',')

# fill empty cells
dataset = dataset.fillna('NaN')

# remove uselss columns
dataset.drop(['incident_url','source_url', 'incident_url_fields_missing', 'sources'], axis=1, inplace=True)

print('klaar')
```

```python
#############
# NIEUW: met alle namen
############
# transforms values to dics
def todict(data_class):
    for word in data_class:
        # selecteer dataset
        column = dataset[word]
        counter = 0 

        # loop door rijen van huidige kolom
        for row in column:   

            # als rij niet leeg is (soms is ie leeg)
            if not row == 'NaN':
                # maak lege dict
                splitted_dict = {}
                
                # als dubbele bar in df
                if '||' in row:
                    cleanrow = row.split("||")
                    for element in cleanrow:
                        index, status = element.split("::")
                        splitted_dict[index] = status
                        
                # als enkele bar in df        
                elif '|' in row:
                    cleanrow = row.split('|')
                    for element in cleanrow:
                        index, status = element.split(":")
                        splitted_dict[index] = status

                # als maar 1 naam
                else:
                    cleanrow = row
                    if '::' in row:
                        index, status = cleanrow.split("::")
                        
                    elif ':' in row:
                        index, status = cleanrow.split(":")
                    splitted_dict[index] = status

                dataset.at[counter, word] = splitted_dict
            counter += 1
    
todict(['participant_name', 'gun_stolen', 'gun_type', 'participant_age', 'participant_age_group', 'participant_gender', 'participant_status', 'participant_type', 'participant_relationship' ])

dataset
```
