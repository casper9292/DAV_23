```python
# scatterplot: hoeveel victims bij een suspect, afgezet tegen leeftijd
# aangeven: aantal vicitms killed
# dus df: id, age_suspect, gender_suspect, number_of_victims, number_of_kills

#participant status: Unharmed, Killed, Injured, else Unknown
 
# make column names
plist = [['incident_id', 'gender_suspect', 'age_suspect', 'victims', 'killed', 'injured', 'unharmed', 'suspectcount','police_suspect', 'police_victim', 'p_total']]

# set starting values
lastid = 461105
victimcount = 0
suspectcount = 0
killedcount = 0
injuredcount = 0
unharmedcount = 0
suspectlist = []
police_suspect = 0
police_victim = 0
p_total = 0


for index, row in df_p.iterrows():
    currentid = row['incident_id']
    if currentid != lastid:
        for suspect in suspectlist: 
            # sla alles op in list 
            plist.append([lastid, suspect[0], suspect[1], victimcount, killedcount, injuredcount, unharmedcount, suspectcount, police_suspect, police_victim, p_total])
        
#         print(plist)
        # reset variables:
        victimcount = 0
        suspectcount = 0
        killedcount = 0
        injuredcount = 0
        unharmedcount = 0
        suspectlist = []
        police_suspect = 0
        police_victim = 0
        p_total = 0

        
    if row['participant_type'] == 'Victim':
        victimcount += 1
        if 'Killed' in row['participant_status']:
            killedcount += 1
            
        if 'Injured' in row['participant_status']:
            injuredcount += 1
            
        if 'Unharmed' in row['participant_status']:
            unharmedcount += 1
            
            
    elif row['participant_type'] == 'Subject-Suspect':
        suspectcount+= 1
        gender_suspect = row['participant_gender']
        age_suspect = row['participant_age']
        
        suspectlist.append([gender_suspect, age_suspect])
        
    if 'officer' in row['participant_name'].lower():
        if row['participant_type'] == 'Subject-Suspect':
            police_suspect += 1
        elif row['participant_type'] == 'Victim':
            police_victim += 1

    elif 'agent' in row['participant_name'].lower():
        if row['participant_type'] == 'Subject-Suspect':
            police_suspect += 1
        elif row['participant_type'] == 'Victim':
            police_victim += 1
    elif 'ofc' in row['participant_name'].lower():
        if row['participant_type'] == 'Subject-Suspect':
            police_suspect += 1
        elif row['participant_type'] == 'Victim':
            police_victim += 1
    elif 'deputy' in row['participant_name'].lower():
        if row['participant_type'] == 'Subject-Suspect':
            police_suspect += 1
        elif row['participant_type'] == 'Victim':
            police_victim += 1
    elif 'swat' in row['participant_name'].lower():
        if row['participant_type'] == 'Subject-Suspect':
            police_suspect += 1
        elif row['participant_type'] == 'Victim':
            police_victim += 1
            
    p_total = police_suspect + police_victim
        
    
    lastid = currentid

    
print('done')
```

```python
# number of police VICTIM incidents per year
vict_states = {}
for index, row in df_merged_police.iterrows():
    if row['police_victim'] != 0:
        
        key = row['state']
        if key in vict_states:
            vict_states[key] +=1
        else:
            vict_states[key] = 1

vict_states['Rhode Island'] = 0
print(vict_states)


# number of police VICTIM incidents per year
suspect_states = {}
for index, row in df_merged_police.iterrows():
    if row['police_suspect'] != 0:
        key = row['state']
        if key in suspect_states:
            suspect_states[key] +=1
        else:
            suspect_states[key] = 1
        
suspect_states['Alaska'] = 0
suspect_states['District of Columbia'] = 0
suspect_states['Maine'] = 0
suspect_states['New Hampshire'] = 0
suspect_states['North Dakota'] = 0
suspect_states['Rhode Island'] = 0
suspect_states['Utah'] = 0
suspect_states['Vermont'] = 0
suspect_states['Wyoming'] = 0
print(suspect_states)

df_vict_states = pd.DataFrame.from_dict(vict_states, orient='index')
df_vict_states['state'] = df_vict_states.index
df_vict_states = df_vict_states.rename(columns={0: 'p_victim'})
df_vict_states = df_vict_states.sort_values(by=['state'])
df_vict_states = df_vict_states.reset_index(drop=True)

df_suspect_states = pd.DataFrame.from_dict(suspect_states, orient='index')
df_suspect_states['state'] = df_suspect_states.index
df_suspect_states = df_suspect_states.rename(columns={0: 'p_suspect'})
df_suspect_states = df_suspect_states.sort_values(by=['state'])
df_suspect_states = df_suspect_states.reset_index(drop=True)
# display(df_suspect_states)
```

```python
# df met aantal inwoners per staat
###
# read population csv
###
df_pop = pd.read_csv("popcount.csv", sep=';')
df_pop = df_pop.drop(51)

df_police_capita = df_pop 
df_police_capita['p_victim'] = df_vict_states['p_victim']
df_police_capita['p_suspect'] = df_suspect_states['p_suspect']

# df_police_capita['capita_p_victim']

capita_p_victim = []
capita_p_suspect = []
for index, row in df_police_capita.iterrows():
    number = row['2015'].replace(".", "")
    capita_p_victim.append((row['p_victim']/float(number))*100000)
    capita_p_suspect.append((row['p_suspect']/float(number))*100000)

df_police_capita['capita_p_victim'] = capita_p_victim
df_police_capita['capita_p_suspect'] = capita_p_suspect

df_police_capita
# moet rhode island weg?
# df_police_capita = df_police_capita.drop(39)
```
