```python
# get df with incidents in each year
def get_inc(dataset, datayear):
    datalist = []
    # get column names
    cols = list(dataset)

    for index, row in dataset.iterrows():
        if row.date.year == datayear:
            datalist.append(dict(zip(cols, row)))
            
    return(pd.DataFrame(datalist))

# get number of incidents in state of each year
def get_inc_states(dataframe_year, currentstate):
    statelist = dataframe_year['state'].tolist()
    if currentstate in statelist:
        return(dataframe_year['state'].value_counts()[currentstate])
    else: 
        return('NaN')

# get number of inhabitants of state in year
def get_pop_count(dataset, pop_state, pop_year):
    df_state = df_pop.loc[df_pop['State'] == pop_state, pop_year]
    return(df_state.iloc[0])

# get number of incident in year and divide over population
def get_inc_per_pop(state, year):
    incidents_year = get_inc(df, year)
#     print('inc per year : {}'.format(incidents_year))
    incidents_state_year = get_inc_states(incidents_year, state)
#     print('inc state per year : {}'.format(incidents_state_year))
    if incidents_state_year != 'NaN':  
        pop_count = get_pop_count(df_pop, state, str(year))
        return(int(incidents_state_year)/int(pop_count.replace('.', '')))
    else:
        return('NaN')

# print((dataframe_year['state'].value_counts()[test_inc_per_pop]))
```

```python
from bokeh.core.properties import value
from bokeh.io import show, output_file
from bokeh.models import ColumnDataSource
from bokeh.plotting import figure

output_file("stacked.html")

state = df_stateyears['state'].tolist()
years = ['2014', '2015', '2016', '2017']
colors = ["#77c298", "#c9d9d3", "#718dbf", "#e84d60"]

data = {'state' : state,
        '2014'   : (df_stateyears['2014']*100000).tolist(),
        '2015'   : (df_stateyears['2015']*100000).tolist(),
        '2016'   : (df_stateyears['2016']*100000).tolist(),
        '2017'   : (df_stateyears['2017']*100000).tolist()}

source = ColumnDataSource(data=data)

p = figure(x_range=state, plot_height=700, plot_width=1200, title="State Counts by Year - incidents per 100.000 inhabitants",
           toolbar_location=None, tools="")

p.vbar_stack(years, x='state', width=0.8, color=colors, source=source,
             legend=[value(x) for x in years])

p.xaxis.major_label_orientation = 1.2
p.y_range.start = 0
p.x_range.range_padding = 0.1
p.xgrid.grid_line_color = None
p.axis.minor_tick_line_color = None
p.outline_line_color = None
p.legend.location = "top_left"
p.legend.orientation = "horizontal"

# p.title.text_font_size="25px"
# p.title.align="center"

p.xaxis.axis_label="State"
p.yaxis.axis_label="incidents per 100.000 inhabitants"

show(p)
output_file('plot_states_stacked.html')
save(p)

```
