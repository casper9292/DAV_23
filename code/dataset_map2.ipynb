# Map per year

```python
# read df met alle incidents per jaar
df_stateyears = pd.read_csv('stateyears.csv')

# sort df
df_stateyears = df_stateyears.sort_values(by=['state'])
df_stateyears = df_stateyears.reset_index(drop=True)

# add centroids lon lat
df_stateyears['lon'] = df_lonlats['longitude']
df_stateyears['lat'] = df_lonlats['latitude']

# add average column to stateyears
df_stateyears['capita'] = df_stateyears[['2014', '2015', '2016', '2017']].mean(axis=1)
df_stateyears['capsize'] = df_stateyears.loc[:, 'capita'] * 100000

df_stateyears
```

```python
df_stateyears_dict = df_stateyears.drop(['Unnamed: 0', 'total number of incidents', 'lon', 'lat'], axis=1)

# multiply x 100.000 all years
df_stateyears_dict.loc[:,'2013'] *= 100000
df_stateyears_dict.loc[:,'2014'] *= 100000
df_stateyears_dict.loc[:,'2015'] *= 100000
df_stateyears_dict.loc[:,'2016'] *= 100000
df_stateyears_dict.loc[:,'2017'] *= 100000

df_stateyears_dict = df_stateyears_dict.set_index("state", drop=True)
df_stateyears_dict = df_stateyears_dict.to_dict(orient="index")


display(df_stateyears_dict['Alabama']['2015'])
```

```python
### plot per year

from bokeh.io import show
from bokeh.models import (
    ColumnDataSource,
    HoverTool,
    LogColorMapper
)
# from bokeh.palettes import Viridis6 as palette
from bokeh.palettes import Viridis256 as palette
from bokeh.plotting import figure

from bokeh.sampledata.us_states import data as states

palette.reverse()

# zo zijn variables veranderd:
# unemployment = df_stateyears_dict
# counties = states
states = {
    code: state for code, state in states.items()
}

state_xs = [state["lons"] for state in states.values()]
state_ys = [state["lats"] for state in states.values()]

state_names = [state['name'] for state in states.values()]

# state_rates wordt gemaakt, is lijst met rates behorend bij state, in zelfde volgorde als state_names
year = '2017'
state_rates = []
for name in state_names:
    state_rates.append(df_stateyears_dict[name][year])
    
    
# color mapper - klopt nu nog niet. moet miss niet logaritmisch?
color_mapper = LogColorMapper(palette=palette)

source = ColumnDataSource(data=dict(
    x=state_xs,
    y=state_ys,
    name=state_names,
    rate=state_rates,
))

TOOLS = "pan,wheel_zoom,reset,hover,save"

p = figure(
    tools=TOOLS,
    plot_height=400,
    x_range = (-130, -60), 
    y_range = (23, 52),
    x_axis_location=None, y_axis_location=None
)

p.title.text="Number of incidents per capita"
p.title.text_font_size="25px"
p.title.align="center"

p.grid.grid_line_color = None

p.patches('x', 'y', source=source,
          fill_color={'field': 'rate', 'transform': color_mapper},
          fill_alpha=0.7, line_color="white", line_width=0.5)

hover = p.select_one(HoverTool)
hover.point_policy = "follow_mouse"
hover.tooltips = [
    ("Name", "@name"),
    ("Incidencts per 100.000 inhabitants", "@rate"),
]


output_file("gmap_filledstates_"+year+".html")
show(p)
```

```python
#######
# RADIOBOX met Jaren WERKEND - allleen nog kleuren fixen!
#####

from bokeh.io import show
from bokeh.models import ColumnDataSource, HoverTool, LogColorMapper, CategoricalColorMapper
from bokeh.plotting import figure
from bokeh.models.widgets import RadioButtonGroup
from bokeh.layouts import column, widgetbox
from bokeh.models.callbacks import CustomJS

# from bokeh.palettes import Viridis6 as palette
from bokeh.palettes import Viridis256 as palette
palette.reverse()

from bokeh.sampledata.us_states import data as states

output_file("plot_states_year_radiobuttons.html")

state_xs = [state["lons"] for state in states.values()]
state_ys = [state["lats"] for state in states.values()]

state_names = [state['name'] for state in states.values()]

# state_rates wordt gemaakt, is lijst met rates behorend bij state, in zelfde volgorde als state_names

def get_rates(year):
    state_rates = []
    for name in state_names:
        state_rates.append(df_stateyears_dict[name][year])
    return(state_rates)

rate_2014 = get_rates('2014')
rate_2015 = get_rates('2015')
rate_2016 = get_rates('2016')
rate_2017 = get_rates('2017')
    
# color mapper - klopt nu nog niet. moet miss niet logaritmisch?
color_mapper = LogColorMapper(palette=palette)
# color_mapper = CategoricalColorMapper(palette=palette, factors=['rate'])

source = ColumnDataSource(data=dict(
    x=state_xs,
    y=state_ys,
    name=state_names,
    rate=rate_2014,
    rate_2014=rate_2014,
    rate_2015=rate_2015,
    rate_2016=rate_2016,
    rate_2017=rate_2017
))

TOOLS = "pan,wheel_zoom,reset,hover,save"

p = figure(
    tools=TOOLS,
    plot_height=400,
    x_range = (-130, -60), 
    y_range = (23, 52),
    x_axis_location=None, y_axis_location=None
)

p.title.text="Number of incidents per capita"
p.title.text_font_size="25px"
p.title.align="center"

p.grid.grid_line_color = None

p.patches('x', 'y', source=source,
          fill_color={'field': 'rate', 'transform': color_mapper},
          fill_alpha=0.7, line_color="white", line_width=0.5)

hover = p.select_one(HoverTool)
hover.point_policy = "follow_mouse"
hover.tooltips = [
    ("Name", "@name"),
    ("Incidencts per 100.000 inhabitants", "@rate"),
]


# add callback to control 
callback = CustomJS(args=dict(p=p, source=source), code="""

            var radio_value = cb_obj.active;
            var data = source.data;            
            x = data['x']
            y = data['y']
            name = data['name']
            rate = data['rate']
            rate_2014 = data['rate_2014']
            rate_2015 = data['rate_2015']
            rate_2016 = data['rate_2016']
            rate_2017 = data['rate_2017']
            
            if (radio_value == 0) {
                for (i = 0; i < x.length; i++) {
                    rate[i] = rate_2014[i]
                }
            }
            
            else if (radio_value == 1) {
                for (i = 0; i < x.length; i++) {
                    rate[i] = rate_2015[i]
                }
            }
            
            else if (radio_value == 2) {
                for (i = 0; i < x.length; i++) {
                    rate[i] = rate_2016[i]
                }
            }
            
            else if (radio_value == 3) {
                for (i = 0; i < x.length; i++) {
                    rate[i] = rate_2017[i]
                }
            }

        source.change.emit();
        """)

# option
option = RadioButtonGroup(labels=["2014", "2015", "2016", '2017'],
                          active=0, callback=callback)
show(column(widgetbox(option),p))

```
