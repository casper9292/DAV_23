```python

import numpy as np
import pandas as pd
import json
import csv

from bokeh.plotting import figure
from bokeh.io import output_file, show
```

```python
# Laad Data
data = pd.read_json("dataset.json", lines=True)
head = pd.read_csv("popcount.csv", sep=';')
```

```python
# Maak dict aan voor states met inwonder
statedict = {}

# Maak lijst met het aantal inwoners per staat en haal de punten weg
headcount = []
for count in head['Census']:
    headcount.append(count)

# Lijst van de Amerikaanse staten
statesofamerica = ["Alabama","Alaska","Arizona","Arkansas","California","Colorado",
  "Connecticut","Delaware","District of Columbia","Florida","Georgia","Hawaii","Idaho","Illinois",
  "Indiana","Iowa","Kansas","Kentucky","Louisiana","Maine","Maryland",
  "Massachusetts","Michigan","Minnesota","Mississippi","Missouri","Montana",
  "Nebraska","Nevada","New Hampshire","New Jersey","New Mexico","New York",
  "North Carolina","North Dakota","Ohio","Oklahoma","Oregon","Pennsylvania",
  "Rhode Island","South Carolina","South Dakota","Tennessee","Texas","Utah",
  "Vermont","Virginia","Washington","West Virginia","Wisconsin","Wyoming", "Puerto Rico"]

# Maak een lijst waarin alle nummers zonder punten worden geschreven
leeg = []

for x in headcount:
    x = x.replace(".", "");
    leeg.append(x)

# Voeg staten toe als key en de inwoners als value
for i in range(len(headcount)):
    x = statesofamerica[i]
    y = leeg[i]
    statedict[x] = y


   
```

```python
# Maak df met alleen suicides

suicides = data.loc[data['incident_characteristics'].str.contains("Suicide")==True]

a = []
b = []
    
# Maak een lijst met staten en incidenten
statecount = suicides['state'].value_counts(sort=False).tolist()
states =  suicides['state'].value_counts(sort=False).index.tolist()

# Maak een dict met de suicide staten en het aantal incidenten
sdict = {}
for i in range(len(statecount)):
    x = states[i]
    y = statecount[i]
    sdict[x]=y

# Zorg dat je het aantal incidenten per 100.000 inwoners krijgt
for x in states: 
    sdict[x] = (sdict[x]/int(statedict[x]))*100000

# Maak lijsten voor de barchart
for y in sdict:
    a.append(y)
    b.append(sdict[y])
```

```python
# Maak barchart

output_file("Suicides_capita.html")

fruits = a

p = figure(x_range=fruits, plot_height=500,plot_width=800, title="Suicides per State Capita",
           toolbar_location=None, tools="")

p.vbar(x=fruits, top=b, width=0.9)

p.xgrid.grid_line_color = None
p.y_range.start = 0

p.xaxis.major_label_orientation = 'vertical'


p.xaxis.axis_label = "State"

p.xaxis.axis_label_standoff = 30

p.yaxis.axis_label = "Number of Suicides per 100.000 citizens"


show(p)
```
